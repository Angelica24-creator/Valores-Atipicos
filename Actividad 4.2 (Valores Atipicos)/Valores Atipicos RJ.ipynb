**DESVIACIÓN ESTANDAR**

```python
#Encuentro los valores extremos 

y=cuantitativos
Limite_Superior=y.mean() +3*y.std()
Limite_Inferior=y.mean() -3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Los outliers se convierten en valores nulos

data2=cuantitativos[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos ua copia del dataframe 

data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

**CUARTILES**

```python
#Metodo aplicando Cuartiles, encuentro cuartiles 0.25 y 0.75

y=cuantitativos

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr=percentile75+1.5*iqr
Limite_Inferior_iqr=percentile25-1.5*iqr
print("Limite superiro permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr) 
```

```python
#Obtenemos datos limpios del DataFrame 

data3_iqr = cuantitativos[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Realizamos ua copia del dataframe 

data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo 

Datos_limpios = pd.concat([cualitativos, data4_iqr], axis=1)
Datos_limpios
```

```python
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertimos dataframe a csv

Datos_limpios.to_csv("Rio de Janeiro sin atipicos.csv")
```
